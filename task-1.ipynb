creating a scaltterplot to viualize sales and profit in my data


```python
sales[["Profit", "Sales"]].plot(kind = "scatter", x = "Profit", y = "Sales")
plt.show()
```

Now creating a box to indenfity outliners

```python
sales[["Profit", "Sales"]].plot(kind = "box", subplots=True)
plt.show()
```

Now creating histograms 

```python
sales["Profit"].plot(kind="hist", bins= 20,title="Profit Distribution")
plt.xlabel("Profit")
plt.show()
```

```python
sns.histplot(data=sales, x="Profit", hue="Region", kde=True, bins=20)
plt.title("Profit Distribution by Region")
plt.show()
```
